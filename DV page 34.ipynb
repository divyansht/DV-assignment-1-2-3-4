```python
import pandas as pd
Location = "D2.csv"
df = pd.read_csv(Location)
df.head()
```

```python
# Create the bin dividers
bins = [0, 60, 70, 80, 90, 120]
# Create names for the four groups
group_names = ['F', 'D', 'C', 'B', 'A']
```

```python
df['lettergrade'] = pd.cut(df['COUNTY'], bins,
 labels=group_names)
df
```

```python
pd.value_counts(df['lettergrade'])
```
