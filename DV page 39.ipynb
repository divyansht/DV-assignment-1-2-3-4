```python
import pandas as pd
Location = "D2.csv"
df = pd.read_csv(Location)
df.head()
```

```python
import numpy as np
df['isFailing'] = np.where(df['COUNTY']<70,
'yes', 'no')
df.tail(10)
```

```python
df['isFailingMale'] = np.where((df['COUNTY']<70)
 & (df['NAME'] == 'Sarasota County'),'yes', 'no')
df.tail(10)
```
